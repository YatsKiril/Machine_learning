15)Розв'язати систему лінійних рівнянь A * x = b, де A - матриця коефієнтів, b - вектор вільних членів

```python
a = np.array([[1,3,-4], [0,2,-3], [-2,2,5]])
b = np.array([5,3,4])

x = np.linalg.solve(a,b)
print(x)

inv_a = np.linalg.inv(a)
x1 = inv_a.dot(b)
print(x1)
```

22)Сортування рядків або стовпців матриці

```python
b = np.matrix([[5,3,4], [2,3,5], [3,2,3]])

print(b)
print(np.sort(b, axis=0))
print(np.sort(b, axis=1))
print(np.sort(b, axis=None))
```

```python
a = np.array([[1,4], [3,1]])

print(a)
print(np.sort(a))
```
